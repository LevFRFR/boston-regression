## Exploring the Data

```python
import pickle

path = '/content/drive/MyDrive/Colab Notebooks/ml project #2 2023-02-24/'

# load the train and test sets from the binary files
with open(path + 'boston_trainSet.pkl', 'rb') as f:
    train_set = pickle.load(f)

with open(path + 'boston_testSet.pkl', 'rb') as f:
    test_set = pickle.load(f)
```

```python
boston = train_set.copy()
```

```python
corr_matrix = boston.corr()

corr_matrix['MEDV'].sort_values(ascending=False)

# average number of rooms per dwelling
# and proportion of residential land zones have a
# positive correlation with a median house value

# meanwhile lower status population and pupil-
# teacher ratio have negative correlation
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

atts = ['MEDV', 'RM', 'ZN', 'LSTAT', 'PTRATIO']
atts2 = ['MEDV', 'INDUS', 'AGE', 'CRIM', 'NOX']
atts3 = ['MEDV', 'TAX', 'B']

scatter_matrix(boston[atts3], figsize=(12,8))
plt.show()

# 1st batch of atts
# RM and LSTAT look like the most coherent correlations
# also, LSTAT looks more like a curve, than a line -> polynomial feature?

# 2nd batch of atts
# let's take CRIM too
```

```python
boston.plot(kind='scatter', x='AGE', y='MEDV', alpha=0.3, grid=True)
plt.show()

# seems like MEDV was maxed out at 50K
# as there is a visible horizontal line there
```
